## Esercizio 1: Somma a segni alterni

Scrivere un programma che riceva in input una sequenza di numeri interi, e che calcoli la somma alternata dei suoi elementi.

Ad esempio, se il programma legge i dati `1 4 9 16 9 7 4 9 11`, deve calcolare

\+ 1 – 4 + 9 – 16 + 9 – 7 + 4 – 9 + 11 = -2

e visualizzare:

`La somma alternata è: –2`

````python
def somma_alternata(dati):
    """
    Calcola la somma alternata di una lista di valori.

    :param dati: la lista di valori da elaborare
    :return: la somma alternata dei valori forniti
    """
    totale = 0
    for i in range(len(dati)):
        if i % 2 == 0:  # indici 0, 2, 4, ...
            totale += dati[i]
        else:  # indici 1, 3, 5, ...
            totale -= dati[i]

    return totale

if __name__ == '__main__':

    msg = 'Inserisci un numro oppure premi Invio per terminare: '

    def main():
        # valori = [1, 4, 9, 16, 9, 7, 4, 9, 11]

        # Legge i valori dall'utente e li memorizza in una lista.
        valori = []
        input_str = input('Inserisci un numro: ')
        while input_str != '':
            valori.append(float(input_str))
            input_str = input('Inserisci un altro numro oppure premi solo Invio '
                              'per terminare: ')
        
        # Mostra la somma alternata dei valori inseriti.
        print('La somma alternata è:', somma_alternata(valori))

        # Shortcut: usa la funzione 'sum' sulla fetta composta dagli elementi
        # in posizione pari e dispari separatamente.
        print(
            'Un altro modo per calcolare la somma alternata:',
            sum(valori[0::2]) - sum(valori[1::2])
        )


    # Chiama la funzione principale.
    main()
````

````python
from random import randint

# Definisci costanti.
NUM_INTERI = 10
VALORE_MAX = 100

# Genera i numeri interi casuali da 1 a VALORE_MAX
valori = []
for i in range(0, NUM_INTERI):
    valori.append(randint(1, VALORE_MAX))

# -----------------------------------------------
# I. Visualizza gli elementi in posizione pari+

# Alternativa 1: utilizza un ciclo for con range a step (passo) 2 
print('Elementi in posizione pari: ', end='')
for i in range(0, len(valori), 2):
    print(valori[i], end=' ')
print()

# Alternativa 2: utilizza lo slicing con step (passo) 2
print('Elementi in posizione pari: ', end='')
for val in valori[::2]:
    print(val, end=' ')
print()


# -----------------------------------------------
# II. Visualizza gli elementi con valore pari

# Alternativa 1: utilizza un ciclo for..range
print('Gli elementi pari sono: ', end='')
for i in range(len(valori)):
    if valori[i] % 2 == 0:
        print(valori[i], end=' ')
print()

# Alternativa 2: utilizza un ciclo for..in
print('Gli elementi pari sono: ', end='')
for num in valori:
    if num % 2 == 0:
        print(num, end=' ')
print()


# -----------------------------------------------
# III. Visualizza gli elementi in ordine inverso.

print('In ordine inverso: ', end='')
for i in range(len(valori) - 1, -1, -1):
    print(valori[i], end=' ')
print()

# Alternativa (invertendo la lista con una porzione)
print('In ordine inverso: ', end='')
for val in valori[::-1]:
    print(val, end=' ')
print()


# -----------------------------------------------
# IV. Visualizza il primo e l'ultimo elemento
print('Primo e ultimo:', valori[0], valori[len(valori) - 1])
print('Primo e ultimo:', valori[0], valori[-1])  # scorciatoia: indice negativo
````

## Esercizio 3: Rimuovere il valore minimo

Scrivere una funzione `rimuovi_minimo(v)` che rimuova la prima occorrenza del valore minimo da una lista `v` senza usare la funzione `min()` né il metodo `remove()`.

La funzione deve restituire una nuova lista, lasciando inalterata quella originale.

````python
from random import randint

def main():
    # Crea una lista di valori casuali.
    valori_casuali = []
    for i in range(10):
        valori_casuali.append(randint(0, 10))

    # Consente all'utente di confrontare le due liste visivamente.
    print('La lista originale è: ', valori_casuali)
    valori_casuali = rimuovi_minimo_con_slice(valori_casuali)
    print('Con il minimo rimosso: ', valori_casuali)


def rimuovi_minimo_con_slice(dati):
    """
    Rimuove la prima occorrenza del valore minimo di una lista
    :param dati: la lista di valori da elaborare
    :return: una nuova copia della lista con l'elemento minimo rimosso
    """
    minimo = dati[0]
    posizione_minimo = 0

    for i in range(1, len(dati)):
        if dati[i] < minimo:
            minimo = dati[i]
            posizione_minimo = i

    # Crea una nuova lista escludendo l'elemento minimo.
    nuova_lista = dati[:posizione_minimo] + dati[posizione_minimo + 1:]
    return nuova_lista


# METODI ALTERNATIVI:

def rimuovi_minimo_con_append(dati):
    if not dati:  # Verifica che la lista non sia vuota
        return dati

    minimo = dati[0]
    posizione_minimo = 0

    # Trova il valore minimo e la sua posizione.
    for i in range(1, len(dati)):
        if dati[i] < minimo:
            minimo = dati[i]
            posizione_minimo = i

    # Crea una nuova lista escludendo l'elemento minimo.
    nuova_lista = []
    for i in range(len(dati)):
        if i != posizione_minimo:
            nuova_lista.append(dati[i])

    return nuova_lista


def rimuovi_minimo_con_remove(dati):
    # Crea una copia della lista originale per mantenere i dati inalterati.
    nuova_lista = dati[:]
    
    # Trova il valore minimo nella lista.
    minimo = min(nuova_lista)
    
    # Rimuove la prima occorrenza del valore minimo dalla copia della lista.
    nuova_lista.remove(minimo)
    
    return nuova_lista

# Chiama la funzione principale.
main()
````

## Esercizio 4: Gli stessi elementi

Scrivere la funzione `same_set(list_a, list_b)` che verifichi se due liste contengono gli stessi elementi, indipendentemente dall’ordine e ignorando la presenza di duplicati. Ad esempio, le due liste `1 4 9 16 9 7 4 9 11` e `11 11 7 9 16 4 1` devono essere considerate uguali.

La funzione non deve modificare le liste che sono state passate come parametri.


````python
# Definisci le costanti.
LISTA_1 = [1, 4, 9, 16, 9, 7, 4, 9, 11]
LISTA_2 = [11, 11, 7, 9, 16, 4, 1]


def main():
    print('Lista 1 è', LISTA_1)
    print('Lista 2 è', LISTA_2)
    print('Le liste contengono gli stessi elementi:', stessi_elementi(LISTA_1, LISTA_2))
    print('Le liste contengono gli stessi elementi:', stessi_elementi_alternativa(LISTA_1, LISTA_2))


def stessi_elementi(list_1, list_2):
    """
    Determina se due liste contengono gli stessi elementi in qualsiasi ordine,
    ignorando eventuali duplicati

    :param l1: la prima lista da considerare
    :param l2: la seconda lista da considerare
    :return: True se le liste contengono gli stessi elementi, False altrimenti
    """
    for valore in list_1:
        if valore not in list_2:
            return False

    for valore in list_2:
        if valore not in list_1:
            return False

    return True


def stessi_elementi_alternativa(list_1, list_2):
    if set(list_1) == set(list_2):
        return True
    else:
        return False


# Chiama la funzione principale.
main()
````

## Esercizio 6: Levigare gli errori (+)

Spesso i dati raccolti durante un esperimento vanno elaborati per rimuovere parte del rumore di misura.

Uno tra gli approcci più semplici a questo problema prevede di sostituire, in una lista di valori, ciascun valore con la media tra il valore stesso e i due valori adiacenti. Se il valore in esame si trova a una delle due estremità della lista, la media sarà con l’unico valore adiacente.

Per comprendere meglio a cosa possa servire tale operazione, ecco un esempio:

![es_00.07_smooth.png](../../imgs/es_00.07_smooth.png)

Scrivere un programma che svolga tale operazione, creando una nuova lista.

Poi prova senza creare una seconda lista, il che è un po' più compolicato perché bisogna evitare di calcolare la media usando i valore già precedentemente modificati.


````python
"""
Rendere più uniformi i valori in una lista facendo la media con i valori adiacenti.
"""

def main():
    valori = [1, 2, 3, 5, 3, 1, 4]
    copia_valori = list(valori)

    # Visualizza i valori originali.
    print('I valori originali sono:', valori)

    # Rendi uniformi i valori e visualizza il risultato.
    rendi_uniformi(valori)
    print('I valori uniformi sono:', valori)

    valori_uniformi_alternativi = rendi_uniformi_su_nuova_lista(copia_valori)
    print('I valori uniformi sono:', valori_uniformi_alternativi)


def rendi_uniformi(dati):
    """
    Rende uniformi i valori in una lista facendo la media del valore di ogni elemento con i suoi vicini.

    :param dati: dati la lista di valori da uniformare
    :return: None (modifica la lista in-place)
    """
    if len(dati) <= 1:
        return

    # Gestisci il primo elemento nella lista.
    vecchio_sinistro = dati[0]
    dati[0] = (dati[0] + dati[1]) / 2

    # Gestisci tutti i valori tranne il primo e l'ultimo.
    for i in range(1, len(dati) - 1):
        corrente = dati[i]
        dati[i] = (vecchio_sinistro + dati[i] + dati[i + 1]) / 3
        vecchio_sinistro = corrente

    # Gestisci l'ultimo elemento nella lista.
    dati[len(dati) - 1] = (vecchio_sinistro + dati[len(dati) - 1]) / 2


def rendi_uniformi_su_nuova_lista(dati):
    """
    Questa versione crea una NUOVA lista anziché modificare quella corrente, e la restituisce
    """
    risultato = []

    # elemento 0: media degli elementi 0 e 1
    risultato.append((dati[0] + dati[1]) / 2)

    # elementi i = 1...len-1: media di 3 elementi
    for i in range(1, len(dati) - 1):
        valore = (dati[i - 1] + dati[i] + dati[i + 1]) / 3
        risultato.append(valore)

    # ultimo elemento: media di 2
    risultato.append((dati[-2] + dati[-1]) / 2)

    return risultato


# Chiama la funzione principale.
main()
````
