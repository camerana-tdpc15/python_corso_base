## Esercizio - Festone di laurea

Requisiti: insiemi

 
Evviva! Ci siamo laureati! E per coincidenza ci siamo riusciti nella stessa sessione dei nostri 
grandi amici Gianni e Giulia. Ora naturalmente stiamo cercando di organizzare una festa e ci piacerebbe 
farla tutti insieme. 

Purtroppo però a causa delle norme Covid non possiamo fare assembramento e 
sono perciò vietate i raduni con più di **13** persone.

Siccome ora siamo dottori, l'idea è quella di risolvere il problema con un bel programmino Python. Abbiamo tre liste di invitati e siccome noi festeggiati siamo tutti amici ci sono ovviamente persone che verrebbero invitate più volte.

1. Trova il numero di invitati effettivi (le persone che verranno alla festa)
2. Stampa la lista di nomi SENZA duplicati
3. Trova il numero delle persone che hanno ricevuto **almeno** 2 inviti
4. Trova l'elenco delle persone che hanno ricevuto **almeno** 2 inviti


Esempio - dati:

```python
invitati_miei =   ["VittorioG", "LucaB", "DavidL", "GiorgioC", "MichelaF", "GiuliaA", "VittorioG", ]
invitati_gianni = ["SamanthaV", "LucaB", "GiorgioC", "MichelaF", "MartaB", "EmmaK"]
invitati_giulia = ["DavidL", "GiorgioC", "MichelaF", "MassimilianoL", "VittorioG", "RobertoU", "EmmaK"]
```


dopo il tuo codice, deve stampare:

```
Invitati miei:  6
Invitati gianni:  6
Invitati giulia:  7
Numero invitati:  11
Nomi invitati:  {'MassimilianoL', 'MartaB', 'MichelaF', 'EmmaK', 'GiorgioC', 'DavidL', 'VittorioG', 'SamanthaV', 'RobertoU', 'LucaB', 'GiuliaA'}
Numero amici invitati almeno 2 volte: 6
Amici invitati almeno due volte:  {'GiorgioC', 'DavidL', 'MichelaF', 'VittorioG', 'LucaB', 'EmmaK'}
```


````python
# 0.1 creo funzione per verificare se ci sono doppinio in ogni lista
def trova_invitati_ripetuti(nomi):
    # Creo Dizionario per tenere traccia dei conteggi dei nomi
    conteggi = {}
    
    
    # Ciclo for per contare i nomi
    for nome in nomi:
        if nome in conteggi:
            # Incrementa il conteggio se il nome è già presente nel dizionario
            conteggi[nome] += 1 
             
        else:
            # Aggiungi il nome al dizionario se è la prima volta che lo incontri
            conteggi[nome] = 1  
            
    # Verifica se ci sono nomi ripetuti
    nomi_ripetuti = [nome for nome, conteggio in conteggi.items() if conteggio > 1]
    return nomi_ripetuti


invitati_miei =   ["VittorioG", "LucaB", "DavidL", "GiorgioC", "MichelaF", "GiuliaA", "VittorioG", ]
invitati_gianni = ["SamanthaV", "LucaB", "GiorgioC", "MichelaF", "MartaB", "EmmaK"]
invitati_giulia = ["DavidL", "GiorgioC", "MichelaF", "MassimilianoL", "VittorioG", "RobertoU", "EmmaK"]

#------------------------------------------------------------------------------------------------------
# 1. Trova il numero di invitati effettivi (le persone che verranno alla festa)
#invitato = ''
#VERIFICO SE CI SONO DOPPIONI NELLA LISTA 'invitati_miei'
# creo lista dei nomi_ripetuti_miei 
nomi_ripetuti_miei = trova_invitati_ripetuti(invitati_miei)
# se 'invitato è presente nella lista nomi_ripetuti 
for invitato in invitati_miei :
    if invitato in nomi_ripetuti_miei:        
        invitati_miei_copia = invitati_miei.copy()
print(f"i miei invitati sono: {len(invitati_miei) - len(nomi_ripetuti_miei)}")
for nome in invitati_miei_copia:
    if nome in nomi_ripetuti_miei:
        nomi_ripetuti_miei.remove(nome)  
    
    
else:
    print(f"i miei invitati sono: {len(invitati_miei)}")
print('-' *100)    

#VERIFICO SE CI SONO DOPPIONI NELLA LISTA 'invitati_gianni'
# creo lista dei nomi_ripetuti_gianni
nomi_ripetuti_gianni = trova_invitati_ripetuti(invitati_gianni)
# se 'invitato è presente nella lista nomi_ripetuti
for invitato in invitati_gianni :
    if invitato in nomi_ripetuti_gianni:
        print(f"gli invitati di Gianni sono: {len(invitati_gianni) - len(nomi_ripetuti_gianni)}")
    else:
        print(f"gli invitati di Gianni sono: {len(invitati_gianni)}")
print('-' *100)  
    
#VERIFICO SE CI SONO DOPPIONI NELLA LISTA 'invitati_giulia'
# creo lista dei nomi_ripetuti_giulia 
nomi_ripetuti_giulia = trova_invitati_ripetuti(invitati_giulia)
# se 'invitato è presente nella lista nomi_ripetuti
for invitato in invitati_giulia : 
    if invitato in nomi_ripetuti_giulia:
        print(f"gli invitati di Giulia sono: {len(invitati_giulia) - len(nomi_ripetuti_giulia)}")
    else:
        print(f"gli invitati di Giulia sono: {len(invitati_giulia)}")
print('-' *100)  
    
#------------------------------------------------------------------------------------------------------
# 2. Stampa la lista di nomi SENZA duplicati e il numero
# Unisco le liste usando l'operatore di concatenazione +.
lista_unica = invitati_giulia + invitati_gianni + invitati_miei
# Converti la lista combinata in un insieme (set), poiché gli insiemi non consentono elementi duplicati.
# per poi riconvertirla in lista
lista_pulita = list(set(lista_unica))
# stampo il numero degli invitati della lista_pulita
print(f"Numero invitati: {len(lista_pulita)}")
print('-' *100)  
print("Nomi invitati:")
for nome in lista_pulita:    
    print(nome)
print('-' *100)  

#------------------------------------------------------------------------------------------------------
# per gli ultimi 2 output creo una lista 'list_nomi_almeno_2volte'
list_nomi_almeno_2volte = []
# per ogni nome nella lista_unica
for nome in lista_unica:
    #se il nome è presente 2 volte
    if lista_unica.count(nome) >= 2 :
        # appendilo alla lista
        list_nomi_almeno_2volte.append(nome)
        

        
# 3. Trova il numero delle persone che hanno ricevuto **almeno** 2 inviti
print(f"Numero di amici invitati almeno 2 volte: {len(set(list_nomi_almeno_2volte))}")
print('-' *100)  

# 4. Trova l'elenco delle persone che hanno ricevuto **almeno** 2 inviti
print("Amici invitati almeno due volte:")
for nome in set(list_nomi_almeno_2volte):
    print(nome)
print('-' *100)  

````
